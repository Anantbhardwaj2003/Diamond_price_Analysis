```python
import pandas as pd

df = pd.read_csv('D:/diamonds.csv')
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
# Check for missing values
print(df.isnull().sum())

# Check for duplicates
print(df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()
```

```python
# Analyze the distribution of numerical features
df.describe()
```

```python
# Analyze the distribution of categorical features
for col in ['cut', 'color', 'clarity']:
    print(df[col].value_counts())

# Explore relationships between features
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example: Scatter plot of price vs. carat
plt.figure(figsize=(8, 6))
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Price vs. Carat')
plt.show()
```

```python
# Example: Box plot of price by cut
plt.figure(figsize=(8, 6))
sns.boxplot(x='cut', y='price', data=df)
plt.title('Price by Cut')
plt.show()
```

```python
# Explore relationships between features using visualizations

# Price vs. Carat
plt.figure(figsize=(8, 6))
sns.scatterplot(x='carat', y='price', data=df, hue='cut') # Added hue for better visualization
plt.title('Price vs. Carat (Colored by Cut)')
plt.show()
```

```python
# Pairplot to visualize relationships between multiple numerical features
sns.pairplot(df[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']], diag_kind='kde')
plt.show()
```

```python
# Further analysis and visualization can be done based on specific needs and questions.
# Example: Analyzing the distribution of prices for different clarity levels.
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='price', hue='clarity', multiple='stack', kde=True)
plt.title('Price Distribution by Clarity')
plt.show()
```

```python
# More in-depth analysis of numerical features
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
for col in numerical_features:
    print(f"Analysis of {col}:")
    print(df[col].describe())
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)  # KDE for better visualization
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Analyze the relationship between price and other numerical features
for col in numerical_features:
  if col != 'price':
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=col, y='price', data=df)
    plt.title(f'Price vs. {col}')
    plt.show()
```

```python
# Correlation matrix for numerical features
correlation_matrix = df[numerical_features].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# More detailed analysis of categorical features
for col in ['cut', 'color', 'clarity']:
    print(f"\nAnalysis of {col}:")
    print(df[col].value_counts(normalize=True)) # Show proportions
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=df)
    plt.title(f"Counts of {col}")
    plt.show()
```

```python
# Boxplot of price for each category
for col in ['cut', 'color', 'clarity']:
    print(f"\nAnalysis of {col}:")
    print(df[col].value_counts(normalize=True))    
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=col, y='price', data=df)
    plt.title(f"Price Distribution by {col}")
    plt.show()
```

```python
# Example: Violin plot to show distribution of price across different cuts
plt.figure(figsize=(10, 6))
sns.violinplot(x='cut', y='price', data=df)
plt.title("Price Distribution by Cut (Violin Plot)")
plt.show()
```
